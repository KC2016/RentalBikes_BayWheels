```python
df_dez_2019 = df_dez_2019.rename(columns={'start_time':'started_at', 
                                          'end_time':'ended_at', 
                                          'start_station_latitude':'start_lat',
                                          'start_station_longitude':'start_lng',
                                          'end_station_latitude':'end_lat',
                                          'end_station_longitude':'end_lng',
                                          'user_type':'member_casual'})

df_dez_2019 = df_dez_2019.drop('rental_access_method', 1)
df_dez_2019['ride_id']=''
df_dez_2019['rideable_type']=''
df_dez_2019 = df_dez_2019[['ride_id', 'rideable_type', 'started_at', 'ended_at',
       'start_station_name', 'start_station_id', 'end_station_name',
       'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng',
       'member_casual']]
df_dez_2019['member_casual'] = df_dez_2019['member_casual'].replace(['Customer'],'casual')
df_dez_2019['member_casual'] = df_dez_2019['member_casual'].replace(['Subscriber'],'member')

df_dez_2019['start_station_id'] = df_dez_2019['start_station_id'].astype(str)
df_dez_2019['end_station_id'] = df_dez_2019['end_station_id'].astype(str)


df_dez_2019
```

```python
df_dez_2020 = df_dez_2020[['ride_id', 'rideable_type', 'started_at', 'ended_at',
       'start_station_name', 'start_station_id', 'end_station_name',
       'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng',
       'member_casual']]
df_dez_2020
```
