## Data wrangling

```python
# calculating distances of the trips
df['distance'] = df.apply(lambda x: vincenty((x['start_lat'], x['start_lng']), (x['end_lat'], x['end_lng'])).m, axis=1)
```

```python
## another way to calculate the distance

# # calculation the distance start-end
# # testing with the first ride
# loc1=(37.770000,-122.450000)
# loc2=(37.760000,-122.450000)
# hs.haversine(loc1,loc2)

# from haversine import Unit
# #To calculate distance in meters 
# hs.haversine(loc1,loc2,unit=Unit.METERS)
```

```python
# convert columns to timestamp
df.started_at = pd.to_datetime(df.started_at)
df.ended_at = pd.to_datetime(df.ended_at)

# subtract times (end - start)
df['travel_time'] = df.ended_at - df.started_at

df['travel_time'] = df['travel_time']/np.timedelta64(1, 'h')
# df.info()
```

```python
df.head()
```

We can see here some errors here with the end time berfore than the start time.

```python
# drop unnamed column
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
#reorder with distance before member_casual:
df = df[['ride_id', 'rideable_type', 'started_at', 'ended_at', 'travel_time', 
       'start_station_name', 'start_station_id', 'end_station_name',
       'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng',
       'distance', 'member_casual']]
```

## Checking outliers

```python
# check outliers
df.boxplot(grid=False, figsize= (15,8))
```

```python
# find quartiles
cols_outliers = ['travel_time', 'distance']

for col in cols_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    print('Q1 of the column ', col,':', Q1)
    print('Q3 of the column ', col,':', Q3)
    print('IRQ of the column ', col,':', IQR)
    Lower_Whisker = Q1 - 1.5 * IQR
    Upper_Whisker = Q3 + 1.5 * IQR
    print('lower_whisker: ', Lower_Whisker, ',' , 'upper whisker: ', Upper_Whisker)
    #  Outlier Treatment 
    df = df[df[col]< Upper_Whisker]
    df = df[df[col]> Lower_Whisker]
    
```

```python

df['started_at'] = df['started_at'].astype('datetime64[ns]')
print("Original Dataframe:")
df.head()
```

```python
# create a column with weekdays
df['weekday_started'] = df['started_at'].dt.day_name()
#reorder with distance before member_casual:
df = df[['ride_id', 'rideable_type', 'weekday_started', 'started_at', 'ended_at', 'travel_time', 
       'start_station_name', 'start_station_id', 'end_station_name',
       'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng',
       'distance', 'member_casual']]
df.head()
```
